```python
%pylab inline
import librosa
from IPython.display import display,Audio
from numpy.fft import fft
```

```python
#load audio
s, fs = librosa.load("../wavs/Reich_Music18.wav")

timeout = 5
rate = 0.01
d = timeout*rate

t0 = 10

# audio fragment
t1 = int(t0*fs)
t2 = int((t0+d)*fs)
sigin = s[t1:t2]
```

```python
sigout = librosa.effects.time_stretch(sigin,rate)
Audio(data=sigout,rate=fs)
```
